### Definição do Problema


### Análise de Clientes de Shopping Center
**Customer Behavior Analysis**

#### Definição do Problema

#### Empresas de varejo dependem fortemente do entendimento do comportamento de seus clientes para melhorar estratégias de marketing, fidelização e vendas.

Neste projeto foi realizada uma Análise Exploratória de Dados (EDA) sobre um conjunto de dados contendo informações de clientes de um shopping center.

O objetivo da análise é responder perguntas estratégicas como:

    Qual é o perfil demográfico dos clientes?
    Existe relação entre renda e consumo?
    Quais produtos são mais consumidos?
    Clientes com filhos possuem comportamento de compra diferente?
    Quais países apresentam maior volume de compras?

Esses insights podem ajudar empresas a personalizar campanhas de marketing, melhorar segmentação de clientes e aumentar receita.

### Instalando e Carregando os Pacotes

```python
# Imports

import numpy as np 
import pandas as pd 
import datetime as dt
import matplotlib.pyplot as plt 
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
sns.set_style('darkgrid')
sns.set(font_scale = 1.2)
```

```python
dados = pd.read_csv('dados/dataset3.csv')
```

O dataset possui 2240 clientes e 28 variáveis

### Limpeza e Preparação dos Dados

```python
dados.rename({' Income ' : 'Income'}, axis = 1, inplace = True)
```

```python
dados['Income'] = dados['Income'].str.replace('$', '')
```

```python
dados['Income'] = dados['Income'].str.replace(',', '').astype(float)
```

#### Conversão de data

A variável Dt_Customer foi convertida para formato datetime

```python
dados['Dt_Customer'] = pd.to_datetime(dados['Dt_Customer'])
```

```python
# Verificando valores ausentes

dados.isna().sum()
```

#### Tratamento de valores ausentes

Foram encontrados 24 valores nulos na variável Income.

Como representam uma pequena parcela da base, os valores foram substituídos pela média da renda.

```python
dados['Income'] = dados['Income'].fillna(dados['Income'].mean())
```

```python
dados['Customer_Age_When_Enrolled'] = dados['Dt_Customer'].dt.year - dados['Year_Birth']
```

#### Criação de Novas Variaveis

>Foram criadas novas variáveis para enriquecer a análise.

**Total Spent**

Representa o gasto total do cliente em todas as categorias de produtos.

**Total Purchases**

Total de compras realizadas considerando todos os canais.

**Kids_Teen_at_home**

Quantidade de filhos ou adolescentes na residência.

**Customer_Age_When_Enrolled**

Idade do cliente no momento em que se cadastrou.

>Essas variáveis ajudam a compreender melhor o comportamento de consumo.

```python
dados[dados['Customer_Age_When_Enrolled'] >= 80]
```

```python
dados['Total Spent'] = (dados['MntFishProducts'] + dados['MntWines'] + dados['MntSweetProducts']
                       + dados['MntFruits'] + dados['MntMeatProducts'] + dados['MntGoldProds'] )
```

```python
dados['Total Purchases'] = (dados['NumDealsPurchases'] + dados['NumWebPurchases'] + dados['NumStorePurchases'] +
                            dados['NumCatalogPurchases'])
```

```python
dados['Kids_Teen_at_home'] = dados['Kidhome'] + dados['Teenhome']
```

```python
dados = dados.drop(['Kidhome', 'Teenhome'], axis = 1)
```

### Análise Exploratória de Dados

**Distribuição do Ano de Nascimento**

**Dataviz 24 - Violin Plot com a relação entre país e total de compras**

```python
# Dataviz 24
plt.figure(figsize = (12, 7))
sns.violinplot(x = dados['Country'], y = dados['Total Purchases'], palette = 'coolwarm')
```

**Dataviz 25 - Pie Chart do agrupamento da variável de total de compras**

```python
# Dataviz 25
pm = dados[['Total Purchases', 'Country']].groupby(['Country']).agg([sum])

sns.set_palette('Spectral')
plt.figure(figsize = (7, 7))
plt.pie(pm['Total Purchases']['sum'], 
        labels = pm.index, 
        explode = (0, 0.4, 0, 0.3, 1, 0.1, 0.1, 0.5),
        shadow = True, 
        autopct = '%1.1f%%')
plt.show()
```
