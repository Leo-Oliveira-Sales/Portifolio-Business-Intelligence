#### **1. Definição do Problema de Negócio**

#### **1.1. O Problema de Negócio:**

Nossa loja de e-commerce está em fase de crescimento, registrando um volume cada vez maior de transações diárias. No entanto, essa grande quantidade de dados de vendas, em seu estado  bruto,  é  como  um  baú  de  tesouro  trancado:  sabemos  que  há  valor  ali,  mas  não conseguimos acessá-lo.

Atualmente,   muitas   de   nossas   decisões   estratégicas   são   baseadas   em   intuição   e observações parciais, o que nos leva a enfrentar os seguintes desafios:

**Gestão  de  Estoque  Ineficiente:**  
Não  temos  clareza  sobre  quais  produtos  são  nossos "campeões de venda" e quais estão parados nas prateleiras. Isso resulta em excesso de estoque de itens de baixa procura e falta de produtos de alta demanda.

**Marketing com Baixo Retorno:** 
Nossas campanhas de marketing são genéricas, pois não sabemos quais categorias de produtos atraem mais os clientes ou em quais regiões geográficas nosso público está mais concentrado.

**Perda de Oportunidades Sazonais:**
Não conseguimos identificar padrões ou tendências de vendas ao longo dos meses. Isso nos impede de planejar promoções estratégicas para períodos de alta ou de criar ações para impulsionar as vendas em meses de baixa.

**Expansão sem Direção:** 
Temos o desejo de expandir, mas não sabemos quais mercados regionais são mais promissores ou onde nossos esforços logísticos deveriam ser focados.

**O problema central é a falta de visibilidade clara sobre a performance do negócio, o que nos impede de tomar decisões rápidas, inteligentes e baseadas em evidências.**

#### **1.3. Solução Proposta**

*A  solução  consiste  em  consolidar,  limpar  e  analisar  o  histórico  de  dados  de  vendas  da nossa plataforma. Utilizando ferramentas de análise de dados (como Python com Pandas, NumPy),  vamos  processar  essas  informações e  criar  um  relatório que  apresente  as descobertas de forma clara e intuitiva para as equipes de gestão, marketing e operações*

#### **1.4. Resultados Esperados e Benefícios de Negócio**

Ao final deste projeto, esperamos alcançar os seguintes resultados:

**1. Otimização  de  Estoque:**  Com  a  lista  dos  produtos  mais  e  menos  vendidos,  poderemos ajustar nossas compras, reduzir custos com armazenamento e evitar a perda de vendas por falta de produto.

**2. Marketing Direcionado e Eficaz:** Sabendo quais categorias e regiões são mais lucrativas, a equipe  de  marketing  poderá  criar  campanhas  3. segmentadas,  aumentando  o  retorno  sobre  o investimento (ROI).

**3. Planejamento  Estratégico:**  A  visualização  das  tendências  mensais  permitirá  um  melhor planejamento  financeiro,  promocional  e  de  recursos,  antecipando  períodos  de  alta  e  baixa demanda.

**4. Decisões  Baseadas  em  Dados:**  Substituiremos  a  intuição  por  dados  concretos,  criando uma cultura orientada a dados que impulsionará o crescimento sustentável do negócio.

# 5. Limpeza, Pré-processamento e Eng. de Atributos

```python
# Se a coluna 'Data_Pedido' não estiver como tipo datetime, precisamos fazer a conversão
# A coluna pode ser usada para análise temporal

df_vendas['Data_Pedido'] = pd.to_datetime(df_vendas['Data_Pedido'])
```

```python
# Engenharia de atributos
# Criando a coluna 'Faturamento' (preço x quantidade)

df_vendas['Faturamento'] = df_vendas['Preco_Unitario'] * df_vendas['Quantidade']
```

```python
# Engenharia de atributos
# Criar uma coluna de status de entrega
# Se a estrega for no Estados de SP, RJ, MG será rapida

df_vendas['Status_Entrega'] = df_vendas['Estado'].apply(lambda estado: 'Rápida' if estado in ['SP', 'RJ', 'MG'] else 'Normal')
```

```python
df["Mes"] = df["Data_Pedido"].dt.to_period("M")
```

```python
df_vendas.info()
```

```python
df_vendas.head()
```

## 📊 KPIs Principais do Negócio

```python
faturamento_total = df["Faturamento"].sum()
total_pedidos = df["ID_Pedido"].nunique()
ticket_medio = faturamento_total / total_pedidos
clientes_unicos = df["ID_Cliente"].nunique()
```

# 6. Análise 1 - Top 10 produtos mais vendidos

```python
# Agrupa por nome do produto, soma a quantidade e ordena para encontrar os mais vendidos

top_10_produtos = df_vendas.groupby('Nome_Produto')['Quantidade'].sum().sort_values(ascending = False).head(10)
```

```python
top_10_produtos
```

```python
plt.figure(figsize=(10,6))

top_10_produtos.sort_values().plot(kind="barh")

plt.title("Top 10 Produtos Mais Vendidos")
plt.xlabel("Quantidade Vendida")
plt.ylabel("Produto")

plt.grid(axis="x", linestyle="--", alpha=0.6)

plt.show()
```

Produtos como Mouse Vertical e Laptop Gamer apresentam alta demanda e devem receber prioridade em:

* estoque
* campanhas promocionais

# 7. Analise 2 - Faturamento Mensal

**Qual foi o faturamento mensal?**

```python
df_vendas.head()
```

```python
# Agrupa por mês e soma o faturamento
faturamento_mensal = df_vendas.groupby('Mes')['Faturamento'].sum()
```

```python
plt.figure(figsize=(10,6))

faturamento_mensal.plot(marker="o",linewidth=3)

plt.title("Faturamento Mensal")
plt.xlabel("Mês")
plt.ylabel("Faturamento (R$)")

plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

Observa-se crescimento no faturamento até março, indicando possível tendência de aumento na demanda.

Abril apresenta queda por ser mês incompleto no dataset.

```python
# Formata para duas casas decimais
faturamento_mensal.map('R$ {:,.2f}'.format)
```

## 8. Análise 3 - Vendas Por Estado

**Qual o total de vendas por estado?**

```python
# Agrupa por estado e soma o faturamento
vendas_estado = df_vendas.groupby('Estado')['Faturamento'].sum().sort_values(ascending = False)
```

```python
# Formata para duas casas decimais
vendas_estado.map('R$ {:,.2f}'.format)
```

```python
plt.figure(figsize=(10,6))

vendas_estado.plot(kind="bar")

plt.title("Faturamento por Estado")
plt.xlabel("Estado")
plt.ylabel("Faturamento")

plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.show()
```

Estados com maior faturamento:

        RJ
        PR
        SP

Essas regiões devem ser foco de:

    marketing
    logística
    campanhas promocionais

## 9. Análise 4 - Faturamento Por Categoria

**Qual o faturamento total por categoria?**

```python
# Agrupa por categoria, soma o faturamento e formata como moeda para melhor leitura
faturamento_categoria = df_vendas.groupby('Categoria')['Faturamento'].sum().sort_values(ascending = False)
```

```python
# O .map('{:,.2f}'.format) é opcional, mas deixa a visualização do número mais clara
faturamento_categoria.map('R$ {:,.2f}'.format)
```

```python
plt.figure(figsize=(8,8))

plt.pie(faturamento_categoria, labels=faturamento_categoria.index, autopct="%1.1f%%",startangle=90)

plt.title("Participação de Faturamento por Categoria")

plt.show()
```

A categoria Eletrônicos domina o faturamento, representando a maior parte da receita da empresa.

## 10. Análise 5 — Ticket Médio por Categoria

```python
ticket_categoria = (df.groupby("Categoria")["Faturamento"].mean())
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    x=ticket_categoria.index,
    y=ticket_categoria.values
)

plt.title("Ticket Médio por Categoria")

plt.ylabel("Ticket Médio (R$)")

plt.show()
```

Categorias de hardware e eletrônicos possuem ticket médio maior, sendo estratégicas para crescimento de receita.
